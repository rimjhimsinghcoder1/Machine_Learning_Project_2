```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler

# Assuming 'data' is your DataFrame
features = data[['FCVC', 'NCP',   'FAF', 'TUE', 'Age', 'Height', 'Weight']]
target = data['Obesity']

# Data Scaling
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42)

# Create k-NN Classifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)

# Train the Model
knn_classifier.fit(X_train, y_train)

# Make Predictions
predictions = knn_classifier.predict(X_test)

# Evaluate Accuracy
accuracy = accuracy_score(y_test, predictions)
print(f'Accuracy: {accuracy}')

# Display Classification Report
print('Classification Report:')
print(classification_report(y_test, predictions))
```

```python
#The output you provided is from the classification report, which provides various metrics for evaluating the performance of a classification model. Here's how to interpret the key metrics:

#1. **Precision:** Precision is the ratio of correctly predicted positive observations to the total predicted positives. In your case:
   - Precision for class 0 (Normal Weight or Insufficient Weight): 0.95
   - Precision for class 1 (Overweight or Obesity): 0.94

   This means that when the model predicts an instance as belonging to class 0, it is correct 95% of the time, and when it predicts an instance as belonging to class 1, it is correct 94% of the time.

#2. **Recall (Sensitivity):** Recall is the ratio of correctly predicted positive observations to the all observations in actual class. In your case:
   - Recall for class 0: 0.84
   - Recall for class 1: 0.98

   This means that the model correctly identifies 84% of actual instances of class 0 and 98% of actual instances of class 1.

#3. **F1-Score:** The F1-score is the weighted average of precision and recall. It is a balance between precision and recall. In your case:
   - F1-score for class 0: 0.89
   - F1-score for class 1: 0.96

   The F1-score is useful when you want to seek a balance between precision and recall.

#4. **Support:** Support is the number of actual occurrences of the class in the specified dataset. In your case, there are 118 instances of class 0 and 305 instances of class 1.

#5. **Accuracy:** Overall accuracy of the model is 94.33%. This represents the ratio of correctly predicted instances (both true positives and true negatives) to the total number of instances.

#6. **Macro Avg and Weighted Avg:** These are the averages of precision, recall, and F1-score across all classes. Macro avg treats all classes equally, while weighted avg takes into account the class imbalance.

#In summary, your model has achieved high accuracy, and it performs well in both precision and recall for both classes. Class 1 (Overweight or Obesity) has particularly high precision and recall, indicating that the model is effective in identifying instances of this class.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler

# Assuming 'data' is your DataFrame
features = data[['FCVC', 'NCP', 'FAF', 'TUE', 'Age', 'Height', 'Weight']]
target = data['Obesity']

# Data Scaling
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42)

# Create Gaussian Naive Bayes Classifier
naive_bayes_classifier = GaussianNB()

# Train the Model
naive_bayes_classifier.fit(X_train, y_train)

# Make Predictions
predictions = naive_bayes_classifier.predict(X_test)

# Evaluate Accuracy
accuracy = accuracy_score(y_test, predictions)
print(f'Accuracy: {accuracy}')

# Display Classification Report
print('Classification Report:')
print(classification_report(y_test, predictions))
```

```python
# The output you provided is from the evaluation of the Naive Bayes classifier on your test data. Here's an interpretation of the key metrics from the classification report:

  1. **Accuracy (Overall Performance):** The accuracy is approximately 90.3%, indicating that the model correctly predicted the class labels for around 90.3% of the instances in the test set.

2. **Precision:** Precision is a measure of the accuracy of the positive predictions. 
   - For class 0 (0 means Non-obese), precision is 80%. This means that out of all instances predicted as Non-obese, 80% were actually Non-obese.
   - For class 1 (1 means Obese), precision is 95%. This means that out of all instances predicted as Obese, 95% were actually Obese.

3. **Recall (Sensitivity):** Recall is a measure of the ability of the model to capture instances of a certain class.
   - For class 0, recall is 86%. This means that the model captured 86% of all actual Non-obese instances.
   - For class 1, recall is 92%. This means that the model captured 92% of all actual Obese instances.

4. **F1-Score:** The F1-score is the harmonic mean of precision and recall, providing a balance between the two metrics.
   - For class 0, the F1-score is 83%.
   - For class 1, the F1-score is 93%.

5. **Support:** The number of actual instances in each class in the test set.
   - For class 0, there are 118 instances.
   - For class 1, there are 305 instances.

6. **Macro Avg and Weighted Avg:** These are the averages of the metrics across all classes. Macro Avg treats all classes equally, while Weighted Avg accounts for class imbalance by weighting the average based on the number of instances in each class.

In summary, the model seems to perform well, especially in predicting Obese instances (class 1), as indicated by high precision, recall, and F1-score for that class. However, there is some room for improvement in predicting Non-obese instances (class 0), as indicated by slightly lower precision, recall, and F1-score for that class.
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler

# Load the modified data
data = pd.read_csv("obesity_modified.csv")

# Select features and target variable
features = data[['FCVC', 'NCP', 'FAF', 'TUE', 'Age', 'Height', 'Weight']]
target = data['Obesity']

# Data Scaling
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42)

# k-NN Classifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)
knn_classifier.fit(X_train, y_train)
knn_predictions = knn_classifier.predict(X_test)

# Naive Bayes Classifier
naive_bayes_classifier = GaussianNB()
naive_bayes_classifier.fit(X_train, y_train)
nb_predictions = naive_bayes_classifier.predict(X_test)

# Evaluate k-NN Classifier
knn_accuracy = accuracy_score(y_test, knn_predictions)
print(f'k-NN Classifier Accuracy: {knn_accuracy}')
print('k-NN Classifier Classification Report:')
print(classification_report(y_test, knn_predictions))

# Evaluate Naive Bayes Classifier
nb_accuracy = accuracy_score(y_test, nb_predictions)
print(f'Naive Bayes Classifier Accuracy: {nb_accuracy}')
print('Naive Bayes Classifier Classification Report:')
print(classification_report(y_test, nb_predictions))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler

# Load the modified data
data = pd.read_csv("obesity_modified.csv")

# Select features and target variable
features = data[['FCVC', 'NCP', 'FAF', 'TUE', 'Age', 'Height', 'Weight']]
target = data['Obesity']

# Data Scaling
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42)

# k-NN Classifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)
knn_classifier.fit(X_train, y_train)
knn_predictions = knn_classifier.predict(X_test)

# Naive Bayes Classifier
naive_bayes_classifier = GaussianNB()
naive_bayes_classifier.fit(X_train, y_train)
nb_predictions = naive_bayes_classifier.predict(X_test)

# Confusion Matrix for k-NN Classifier
knn_conf_matrix = confusion_matrix(y_test, knn_predictions)
print('Confusion Matrix for k-NN Classifier:')
print(knn_conf_matrix)

# Confusion Matrix for Naive Bayes Classifier
nb_conf_matrix = confusion_matrix(y_test, nb_predictions)
print('Confusion Matrix for Naive Bayes Classifier:')
print(nb_conf_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Heatmap for k-NN Classifier
plt.figure(figsize=(8, 6))
sns.heatmap(knn_conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - k-NN Classifier')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Heatmap for Naive Bayes Classifier
plt.figure(figsize=(8, 6))
sns.heatmap(nb_conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - Naive Bayes Classifier')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# Precision-Recall Curve for k-NN Classifier
knn_precision, knn_recall, _ = precision_recall_curve(y_test, knn_predictions)
plt.plot(knn_recall, knn_precision, marker='.')
plt.title('Precision-Recall Curve - k-NN Classifier')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()

# Precision-Recall Curve for Naive Bayes Classifier
nb_precision, nb_recall, _ = precision_recall_curve(y_test, nb_predictions)
plt.plot(nb_recall, nb_precision, marker='.')
plt.title('Precision-Recall Curve - Naive Bayes Classifier')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()
```
